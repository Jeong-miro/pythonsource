```python
money,coffee = 300,10
#커피가 10개 다 팔리면 돈이 있더라도 팔 지 못하게
# pass : 코딩 미완성 시에도 작동 가능하게 (무한루프)
# break : 반복문 강제종료
while True:
    print('커피 주기')
    coffee = coffee - 1
    print(f"남은 커피 개수 {coffee}")

    if coffee == 0:
        print('커피 판매 종료')
        break
```

```python
# 1~100 사이의 숫자중에서 3의 배수만을 더한 후 결과 값 출력

i = 1
hap = 0
while i <= 100:
    if (i % 3 == 0):
        hap = hap + i
    i = i + 1    
print(f"1~100까지 중 3의 배수의합 : {hap}")
```

```python
# 1~100 사이의 숫자중에서 3의 배수(단 9의 배수 제외)만을 더한 후 결과 값 출력
i = 1
hap = 0
while i <= 100:
    if (i % 3 == 0 and i % 9 != 0):
        hap = hap + i
    i = i + 1    
print(f"1~100까지 중 3의 배수(9의 배수 제외)의합 : {hap}")
```

```python
# 1~100 사이의 숫자중에서 3의 배수와 5의배수만을 더한 후 결과 값 출력

i = 1
hap = 0
while i <= 100:
    if (i % 3 == 0 or i % 5 == 0):
        hap = hap + i
    i = i + 1    
print(f"1~100까지 중 3과 5의 배수의합 : {hap}")
```

```python
# continue : 반복문의 처음으로 다시 돌아가기

i = 0
while i < 10:
    i = i + 1
    if i % 2 == 0 : continue
    print(i)
```

```python
# 주차장 프로그램

# 사용자에게 메뉴를 제공
# [1] 주차 | [2] 출차 | [3] 종료

# 사용자는 메뉴 선택
# 1
# 자동차 A ~ E
# A 자동차 입차. 주차장 상태 ==> ['A'] list
# 1
# B 자동차 입차. 주차장 상태 ==> ['A','B']
# 2 빠져나갈때는 아직 뒤에서만 가능
# B 자동차 출차. 주차장 상태 ==> ['A']
# E 자동차 입차. 주차장 상태 ==> ['A','B','C','D','E']
# 1
# 주차장 만차

parking_lot = []
top, car_name = 0, 'A'
while True:
    no = int(input("[1] 주차 | [2] 출차 | [3] 종료"))
    
    if no <= 3:
        if no == 1:
            if top >= 5:
                print("주차장 만차")
                
            else:
                parking_lot.append(car_name)
                print(f"{car_name}자동차 입차. 주차장 상태 ==> {parking_lot}")
                car_name = chr(ord(car_name)+1) 
                top = top + 1
            
        elif no == 2:
            if top > 0:
                parking_lot.pop()
                print(f"{car_name}자동차 출차. 주차장 상태 ==> {parking_lot}")
                top = top - 1 
                car_name = car_name - 1
            
            else:
                print("빠져나갈 자동차가 없음")
        else:
            print("프로그램 종료")
            break
    else:
        print("번호를 확인해주세요")
```
