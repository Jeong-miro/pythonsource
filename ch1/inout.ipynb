```python
# 1~100 사이의 숫자중에서 3의 배수만을 더한 후 결과 값 출력

i = 1
hap = 0
while i <= 100:
    if (i % 3 == 0):
        hap = hap + i
    i = i + 1    
print(f"1~100까지 중 3의 배수의합 : {hap}")
```

```python
# 1~100 사이의 숫자중에서 3의 배수(단 9의 배수 제외)만을 더한 후 결과 값 출력
i = 1
hap = 0
while i <= 100:
    if (i % 3 == 0 and i % 9 != 0):
        hap = hap + i
    i = i + 1    
print(f"1~100까지 중 3의 배수(9의 배수 제외)의합 : {hap}")
```

```python
# 주차장 프로그램

# 사용자에게 메뉴를 제공
# [1] 주차 | [2] 출차 | [3] 종료

# 사용자는 메뉴 선택
# 1
# 자동차 A ~ E
# A 자동차 입차. 주차장 상태 ==> ['A'] list
# 1
# B 자동차 입차. 주차장 상태 ==> ['A','B']
# 2 빠져나갈때는 아직 뒤에서만 가능
# B 자동차 출차. 주차장 상태 ==> ['A']
# E 자동차 입차. 주차장 상태 ==> ['A','B','C','D','E']
# 1
# 주차장 만차

parking_lot = []
top, car_name = 0, 'A'
while True:
    no = int(input("[1] 주차 | [2] 출차 | [3] 종료"))
    
    if no <= 3:
        if no == 1:
            if top >= 5:
                print("주차장 만차")
                
            else:
                parking_lot.append(car_name)
                print(f"{car_name}자동차 입차. 주차장 상태 ==> {parking_lot}")
                car_name = chr(ord(car_name)+1) 
                top = top + 1
            
        elif no == 2:
            if top > 0:
                outCar = parking_lot.pop()
                print(f"{outCar}자동차 출차. 주차장 상태 ==> {parking_lot}")
                top = top - 1 
                car_name = chr(ord(car_name) - 1)
            
            else:
                print("빠져나갈 자동차가 없음")
        else:
            print("프로그램 종료")
            break
    else:
        print("번호를 확인해주세요")
```

```python
def calculator(num1,num2,op):
    if op == "+":
        return num1+num2

    elif op == '-':
        return num1-num2

    elif op == '*':
        return num1*num2

    elif op == '/':
        return num1/num2

    elif op == '//':
        return num1//num2

# 숫자 2개 입력받기, 연산자 기호 입력받기
# num1 num2 op
num1 = int(input("첫번째 숫자 입력"))
num2 = int(input("두번째 숫자 입력"))
op = input("연산자 기호(+,-,*,/,//)")
print(calculator(num1,num2,op))



```
