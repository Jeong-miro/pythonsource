#### 문자열 포매팅
- 문자열 안에 어떤 값을 삽입하는 방법

| 코드 | 설명 |
| --- | --- |
| `%s`  | 문자열(string) |
| `%d`  | 정수값(integer) |
| `%c`  | 문자 1개(character) |
| `%f`  | 부동 소수(floating-point) |
| `%o`  | 8진수 |
| `%x`  | 16진수 |
| `%%`  | % 문자 자체 |

- format() 함수 사용
- f 문자열 포매팅(파이썬 3.6버전 이상 지원)

```python
# %d => 정수값
a = "I eat %d apples." % 3
print(a)
```

```python
# %s => 어떤 값이든 대입 가능
a = "I eat %s apples." % "five"
print(a)
```

```python
a = "Error is %d%%." % 98
a
```

```python
a = "I ate %d apples. so I was sick for %s days" % (10, "three")
print(a)
```

```python
"%10s" % "hi" # 10자리 공간 확보, 오른쪽 정렬

print("%10s" % "hi")
print("%-10sjane." % "hi") # 왼쪽 정렬

print("%0.4f" % 3.42134234) # 소수점 4자리까지만 표시
print("%10.4f" % 3.42134234) # 전체 길이는 10 / 오른쪽 정렬
#0 : 길이에 상관하지 않음
```

```python
a = "I eat {} apples.".format(5)
a
```

```python
a = "I ate {} apples. so I was sick for {} days".format(10,"three")
a = "I ate {0} apples. so I was sick for {1} days".format(10,"three")
a
```

```python
name = "홍길동"
age = 30
# 포매팅 방법 1 
print("제 이름은 %s 입니다. 제 나이는 %d 입니다." % (name, age))
# 포매팅 방법 2
print("제 이름은 {0} 입니다. 제 나이는 {1} 입니다.".format(name, age))
#f문자열 포매팅
print(f"제 이름은 {name} 입니다. 제 나이는 {age} 입니다.")
```

```python
# TypeError: 'set' object is not subscriptable
# s2[0]
# 인덱싱,슬라이싱 하고 싶다면 가능한 타입으로 변환
# 리스트: list(), 튜플 : tuple()
s1 = "안녕하세요"
print(type(s1)) #str
print(type(3)) #int
print(type(list3)) #list
print(type(t4)) #tuple
print(type(dict2)) #dict
print(type(s2)) #set

#set 자료형 => 새로운 list로 생성
print(list(s2)[:3])
print(tuple(s3))
print(s2)
```

```python
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])

#교집합
print(s1 & s2)
print(s1.intersection(s2))

#합집합
print(s1 | s2)
print(s1.union(s2))

#차집합
print(s1 - s2)
print(s1.difference(s2))
```
