````python
#def add_mul(choice, *args, a):
#    pass

# 일반 매개변수는 앞 쪽으로 선언하는 편
# + * 실행 함수
def add_mul(choice, *args):
    if choice == "add":
        result = 0
        for i in args:
            result = result + i

    elif choice == "mul":
        result = 1
        for i in args:
            result = result * i
    
    return result
````

````python
# add_mul("add",10,11,12,13,a=10)
print(add_mul("add",10,11,12,13))
print(add_mul("mul",10,11,12,13))
````

````python
# 키워드 매개변수 (key=value)
# **변수명

def print_kwargs(**kwargs):
    print(kwargs)

    # key 값만 출력
    for k in kwargs.keys():
        print(k)
        
    # value 값만 출력
    for v in kwargs.values():
        print(v)
    
    # key,value 같이 출력
    for k,v in kwargs.items():
        print(k,v)
````

````python
print_kwargs(name="kim",age=15)
print_kwargs(name="park",age=15, addr="서울")
print_kwargs(name="hong",age=15, addr="경기", tel="010-1234-5678")
````

````python
# 변수 효력 범위
# 1) 리턴이용(추천)

#함수 밖의 변수 : 전역변수 
a = 1
def vartest(a):
    #매개변수 a : 지역변수
    a = a +1
    return a

a = vartest(a)
print(a)
````

````python
# 2) global 키워드 사용

a = 1
def vartest():
    #매개변수 a : 지역변수
    global a
    a = a +1

vartest()
print(a)
````

````python
# 리스트의 평균 구하는 함수
# get_average()

# 리스트의 요소 갯수는 정해지지않음
def get_average(*numbers):
#   print(f"평균 : {sum(numbers)/len(numbers)}")

    result=0
    for i in numbers:
        result = result + i

    print (result / len(numbers))


get_average(10,11,12)
get_average(10,11,12,15,25)
get_average(10,11,12,15,25,36)
````

````python

def calculator(num1,num2,op):
    if op == "+":
        return num1+num2

    elif op == '-':
        return num1-num2

    elif op == '*':
        return num1*num2

    elif op == '/':
        return num1/num2

    elif op == '//':
        return num1//num2


print(calculator(10,20,"+"))
print(calculator(10,20,"-"))
print(calculator(10,20,"*"))
print(calculator(10,20,"/"))
print(calculator(10,20,"//"))
````

````python
#filter()
list1= list(range(1,21))
list1
#filter(함수, iterable)

# 짝수만 새로운 리스트로 생성
#방법 1
list2 = []
for i in list1:
    if i % 2 == 0:
        list2.append(i)
print(list2)
#방법2 (리스트 컴프리헨션)

list3 = [i for i in list1 if i % 2 == 0]
print(list3)
#방법3 filter
# filter(함수, iterable)
list(filter(lambda i:i%2==0, list1))
````

````python
strings = ["Bobs","charles","Alexander3","teddy","apple","Banana"]

# 단어의 시작이 대문자인 단어만 모아서 새로운 리스트 생성

nlist = []
for i in strings:
    if i[0].isupper() == True:
        nlist.append(i)

print(nlist)

nlist2 = [i for i in strings if i[0].isupper() ]
print(nlist2)
#filter(함수,원본리스트) : 새로운 리스트로 생성(함수에 맞춰)

list(filter(lambda i : i[0].isupper() , strings))
````
