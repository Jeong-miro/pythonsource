````python
# 한 시간 임금은 10700원 8시간까지 10700으로 지급
# 만약 8시간 초과 근무시 초과 시간은 1.5배 지급

#근무시간(hours) rate(시간당 임금), pay(일당)
hours,rate,pay = 12, 10700, 0

if(hours > 8):
    overhours = hours - 8
    pay = (hours-overhours) * rate + rate * 1.5 * overhours
    # pay = rate * 8 + ((hours-8) * rate * 1.5)
else:
    pay = hours * rate

print(pay)

print("근무시간 : %d , 일당 : %d" %(hours,pay))
print("근무시간 : {0} , 일당 : {1:.0f}" .format(hours,pay))
print(f"근무시간 : {hours} , 일당 : {pay}")
````

````python
# if~else 를 간략하게 표현
# 점수가 60점 이상이면 성공 아니면 실패
score = 70
if score >= 60 :
    print("success")
else:
    print("fail")

# 조건문 표현식으로 변경

msg = "성공" if score >= 60 else "실패"
print(msg)
print("성공" if score >= 60 else "실패")
````

````python
# 1~100 합계 구하고 출력

i = 1
hap = 0
while i <= 100:
    hap = hap + i
    i = i + 1
print(f"1~100까지의 합 : {hap}")
````
