`숫자가 짝수면 짝수라고 출력, 아니면 홀수라고 출력`

```python
num = 5
# if 문 : bool 자료형
if (num % 2 == 0):
    print("짝수")
else:
    print("홀수")
```

```python
# SyntaxError : 문법 오류

# >, <, >=, <=, ==, != , True, False
# in, not in
a=200
# a가 100보다 작다면 a는 100보다 작다 출력
if a < 100:
    print("a는 100보다 작다")

# a가 100보다 크고 200 이하라면 a는 100과 200 사이의 수 이다 출력

if 100 < a <= 200:
    print("a는 100과 200사이의 수")
```

```python
# 점수가 90이상이면 A , 80이상이면 B, 70 이상이면 C, 나머지는 F
score = 99
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else :
    print("F")
```

```python
a, b, c = 12, 6, 19

# 가장 큰 숫자 출력
if a > b & a > c:
    print("가장 큰 수는 a")
elif b > a & b > c:
    print("가장 큰 수는 b")
else:
    print("가장 큰 수는 c")
```

```python
max = a
if max < b:
    max = b

if max < c:
    max = c
print("가장 큰 수",max)
```

```python
# 가장 작은 수
min = a
if min > b:
    min = b

if min > c:
    min = c

print("가장 작은 수", min)
```

```python
# and : x and y (x,y 가 모두 참인 경우 참)
# or : x or y (x,y 중 하나만 참이여도 참인 경우 참)
# not x : 반전

# score 가 90 이상이고, grade 가 A 인 경우 합격
# 나머지는 불합격

score, grade = 85 , 'A'

if score > 90 and grade == "A":
    print("합격")
else:
    print("불합격")
```
