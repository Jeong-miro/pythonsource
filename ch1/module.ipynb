```python
# random : 난수(규칙이 없는 임의의 수)를 발생시키는 모듈

from random import random, randint, uniform, randrange, choice, shuffle, sample

# random() : 0.0 ~ 1.0 사이의 실수 중에서 난수
# 0.0 <= random.random < 1.0
print(random())
# 시작숫자 <= random.ranint < 끝 숫자
print(randint(1,10))
# uniform : 지정한 두 실수 사이의 난수
print(uniform(10,20))
# randrange(stop)
# randrange(start,stop)
# randrange(start,stop,step)
print(randrange(10)) # 0~9
# choice(시퀀스) : 리스트, 튜플, 문자열 중 택 1
print(choice([1,2,3,4,5,6,7]))
list1 = list(range(1,8))
# shuffle 섞기
shuffle(list1)
print(list1)
# sample(시퀀스,개수) : 리스트, 튜플, 문자열
print(sample(list1, k=2))
```

```python
#loads() : json 문자열 => 파이썬 자료형(dict)로 변환
#dumps() : 파이썬 자료형(dict) => json 문자열로 변환

str1 = """
{
    "name": "홍길동",
    "birth": "0727",
    "age": 20
}
"""

data = json.loads(str1)
print(data)
print(type(data))

print(data['name'])
print(data.get('name'))


```
