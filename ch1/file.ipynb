#### file 읽고 쓰기
| 파일 열기 모드 | 설명 |
|------------|------|
|`r`| **읽기(Read) 모드** : 파일의 내용을 읽기만할때 사용. 파일이 존재하지 않는 경우 오류 발생. |
|`w`| **쓰기(Write) 모드** : 파일의 내용을 새로 쓸때 사용. 기존 파일이 존재하는 경우 내용 삭제 후 새로 생성(덮어쓰기). 없는경우 새로 생성. |
|`a`| **추가(Append) 모드** : 기존 파일의 마지막에 내용을 추가할 때 사용. 파일이 없는경우 새로 생성. |

```python
# score.txt 내용을 읽은 후 평균 구하고 출력
scores = []

with open("./data/score.txt","r",encoding="utf-8") as f:
    for num in f:
        scores.append(int(num))

print(scores)
print(f"평균 : {sum(scores)/len(scores)}")   
```

```python
#sample.txt 파일을 읽어 총합과 평균을 구한 후 총합과 평균을 result.txt 파일에 작성
sample = []

with open("./data/sample.txt","r",encoding="utf-8") as f:
    for num in f:
        sample.append(int(num))

with open("./data/result.txt","w",encoding="utf-8") as f:
    f.write("총합 : %d\n" % sum(sample))
    f.write("평균 : %.2f" % (sum(sample) / len(sample)))
```

```python
# fruits.txt 읽어서
# apple 3
# banana 2
# 과일 단어 몇번 나왔는지 화면에 출력

# 단어 분리 : split() => ['apple', 'banana','orange']

# dict 추가
#if "apple" in dict1:
#    dict1['apple'] += 1
#else:
#    dict1['apple'] = 1
dict1 = {}
with open("./data/fruits.txt", "r", encoding="utf-8") as f:
    for line in f:
        #print(line)
        words = line.split()
        print(words)
        for word in words :
            #print(word)

            if word in dict1:
                dict1[word] += 1
            else:
                dict1[word] = 1

for fruit, count in dict1.items():
    print(f"{fruit} : {count}")
```

```python
# 1000 명의 이름,키,몸무게 info.txt 읽은 후 
# BMI 지수 계산
# bmi = 체중 / 키의 제곱
# bmi 가 18.5 미만이면 저체중
# bmi 가 18.5~22.9 사이면 정상체중
# bmi 가 22.9 초과하면 과체중
# bmi 가 25 이상이면 비만

# 출력문
# 이름 : 하타
# 몸무게 : 95
# 키 : 180
# BMI : 18.6
# 결과 : 정상체중

#name,weight,height = "카사, 69, 156".split(", ")
#print(name,weight,height)
dict1 = {}
with open("./data/info.txt", "r", encoding="utf-8") as f:
    for line in f:
        #print(line)
        name,weight,height = line.strip().split(", ")
        #print(name,weight,height)
        weight = int(weight)
        height = int(height)
        bmi = weight / (height/100)**2
 
        if bmi < 18.5:
            result = "저체중"
        elif bmi < 23:
            result = "정상체중"
        elif bmi < 25:
            result = "과체중"
        else:
            result = "비만"
            
        
        print("\n".join(["이름 : {}", "몸무게 : {}","키 : {}","BMI : {:.1f}","결과 : {}"]).format(name,weight,height,bmi,result))
        print()      
        

```
