- 객체마다 따로따로 가지는 데이터

```python
# 자동차 클래스 설계
# 속성 : 색상, 속도
# 동작 : 전진, 후진, 속도올리기, 속도 내리기...

class Car :
    def __init__(self,color,speed):

        #인스턴스 변수
        self.color = color
        self.speed = speed

        #인스턴스(멤버) 메소드(함수)
    def upSpeed(self, value):
        self.speed += value
        
    def downSpeed(self, value):
        self.speed -= value
        
        # <__main__.Car object at 0x000001719E9E6CF0> => 보고싶은문자로변경
    def __str__(self):
        return f"Car(color = {self.color}, speed = {self.speed})"
    
```

```python
# Account 클래스 작성
# 멤버변수 : 계좌주(owner), 잔액(balance)
#     메소드 : deposit(money) : 잔액 = 잔액 + 입금액
#     메소드 : withdraw(money) : 잔액 = 잔액 - 입금액  #입출금
#     메소드 : showBalance() : 현재 잔액 출력

class Account:
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance

    def deposit(self,money):
       
        self.balance += money

    def withdraw(self,money):
        if self.balance < money:
            print("잔액 부족")
        else:
            self.balance -= money
        
    def showBalance(self):
        print(f"계좌주 : {self.owner} , 현재 남은 잔액 : {self.balance}")

owner1 = Account("홍길동",100000)
owner1.deposit(50000)
owner1.withdraw(60000)
owner1.showBalance()
```

```python
# Order 클래스 작성
# 클래스 변수 : total_order
# 멤버 변수 : 메뉴(menu), 가격(price), 수량(count)
#      메소드 : getTotalPrice() = 가격 * 수량 : 총 주문금액 반환
#              showOrder() : 출력문 작성
#              메뉴 : 아메리카노
#              가격 : 4000원
#              수량 : 2개
#              총 주문금액 : 8000원

class Order:
    def __init__(self,menu,price,count):
        self.menu = menu
        self.price = price
        self.count = count

    def getTotalPrice(self):
        print(f"{self.price * self.count} 원")

    def showOrder(self):
        print(f"메뉴 : {self.menu},\n가격 : {self.price}원,\n수량 : {self.count}개,\n총 주문금액 : {self.price * self.count}원\n")


    @classmethod
    def total_order(cls):
        print(f"최종 주문:{Order.showOrder}")


menu1 = Order("아메리카노",4000,2)
menu2 = Order("케이크",13000,1)

menu1.getTotalPrice()
menu1.showOrder()

menu2.getTotalPrice()
menu2.showOrder()

print(Order.total_order)
```

#### 상속
- 기존 클래스(부모/상위) 의 멤버변수와 메서드를 새로운 클래스(자식/하위)가 그대로 물려받는 것
- 코드 재사용성, 확장

```python
class Parent:
    def __init__(self):
        self.value ="테스트"
        print("Parent 클래스의 __init__() 호출")

    def test(self):
        print("Parent 클래스의 test() 호출")

    


class Child(Parent):
    pass

child = Child()
child.test()
print(child.value)
```

```python
# Audio : 멤버변수 : power(전원), volume(소리조절) / 메소드 : 전원 조절, 볼륨조절 + tune()
# TV : 멤버변수 : power, volume, size / 메소드 : 전원 조절, 볼륨조절 + watch()
# 공통분모 = ower,volume, 전원조절, 볼륨조절

class AudioVisual:
    def __init__(self,power,volume):
        self.power = power
        self.volume = volume

    def switch(self,on_off):
        self.power = on_off

    def setVolume(self, volume):
        self.volume = volume


class Audio(AudioVisual):
    def __init__(self, power, volume):
        super().__init__(power, volume)
    
    def tune(self):
        print("La la la..." if self.power else "turn it on")
    
class TV(AudioVisual):
    def __init__(self, power, volume, size):
        super().__init__(power, volume)
        self.size = size

    
    def watch(self):
        print("have fun!" if self.power else "switch on")

tv = TV(False, 12, 50)
# tv 켜기
tv.switch(True)
# tv 보기
tv.watch()  


audio = Audio(True, 15)
audio.setVolume(10)
audio.tune()
audio.switch(False)
        
```

#### 오버라이딩
- 부모 클래스에 있는 메서드를 자식 클래스에서 같은 이름으로 다시 정의해서 동작 내용 변경
- 

```python
class Animal:
    def __init__(self,name):
        self.name = name

    def sound(self):
        print(f"{self.name}가 소리를 냅니다.")

class Dog(Animal):
    def sound(self):
        print(f"{self.name} : 멍멍!!")
        #return super().sound()

class Cat(Animal):
    def sound(self):
        # 오버라이딩 시 동일한 부모 메소드를 호출하고 싶다면
        super().sound()
        print(f"{self.name} : 야옹!!")
        

dog = Dog("초코")
cat = Cat("몰리")

dog.sound()
cat.sound()
```

```python
# Employee 부모 클래스 :
# 멤버 변수 : 이름(name),급여(pay)
#     메소드 : showInfo() => 이름, 급여 출력

# Manager 자식 클래스
# 멤버 변수 : 보너스(bonus)
#     메소드 : getTotalPay() : 급여 + 보너스 출력

class Employee:
    def __init__(self,name,pay):
        self.name = name
        self.pay = pay

    def showInfo(self):
        return f"이름 : {self.name} 급여 : {self.pay}"
    
class Manager(Employee):
    def __init__(self, name, pay,bonus):
        super().__init__(name, pay)
        self.bonus = bonus
    
    def getTotalPay(self):
        return f"급여:{self.pay} + 보너스:{self.bonus} = {self.pay+self.bonus}"

pay1 = Manager("홍길동",10000,5000)
pay1.getTotalPay()
```

```python
# Shape 부모 클래스
# 메소드 : area() => "도형의 넓이" 출력

# rectangle 자식 클래스
# 멤버 변수 : width, height
# 메서드 : area() => 직사각형의 넓기 출력

# Circle 자식 클래스
# 멤버 변수 : radius(반지름)
# 메서드 : area() => 원의 넓이 출력 

class Shape:
    def area(self):
        print("도형의 넓이")

class rectangle(Shape):
    def __init__(self,width,height):
        super().__init__()
        self.width = width
        self.height = height

    def area(self):
        return f"직사각형의 넓이 : {self.width * self.height}"

class Circle(Shape):
    def __init__(self, radius):
        super().__init__()
        self.radius = radius

    def area(self):
        return f"원의 넓이 : {3.14*self.radius*self.radius}"
a = Shape()
a.area()
b = rectangle(3,5)
print(b.area())
c = Circle(5)
print(c.area())
```
